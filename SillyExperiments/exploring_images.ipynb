```python
# exploring the data


import torch
from torchvision.models import vit_l_16, ViT_L_16_Weights
from torchvision.transforms import Compose
import matplotlib.pyplot as plt

model = vit_l_16(weights=ViT_L_16_Weights.IMAGENET1K_V1)

transforms = ViT_L_16_Weights.IMAGENET1K_V1.transforms()

from PIL import Image
```

```python
# Load your image (replace 'your_image.jpg' with your image path)
image_name_file = 'ADE_train.jpg'
image = Image.open(image_name_file)
print(image.size)
# Apply the transforms
processed_image = transforms(image)

# print(processed_image.shape)
plt.imshow(image)
```

```python


# Remove normalization for visualization
mean = torch.tensor([0.485, 0.456, 0.406])
std = torch.tensor([0.229, 0.224, 0.225])
unnormalized_image = processed_image * std[:, None, None] + mean[:, None, None]

# Convert to numpy and permute dimensions to (H, W, C)
image_np = unnormalized_image.permute(1, 2, 0).clamp(0, 1).numpy()

print(processed_image.shape)
# Display using Matplotlib
plt.imshow(image_np)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
import torchvision.transforms as transforms

image_operation = transforms.Compose([
    transforms.Resize(size=(320,320)),
    transforms.ToTensor()
])

image = Image.open(image_name_file)

new_image = image_operation(image)
print(new_image.shape)
```
